```python
def remove(car):
    
    kmpl, cc, bhp = car
    
    kmpl = kmpl.split(' ')[0]

    cc = cc.split(' ')[0]

    bhp = bhp.split(' ')[0]

    return kmpl,cc,bhp
```

```python
cars[['mileage(kmpl)', 'engine(cc)', 'max_power(bhp)']] = cars[['mileage', 'engine', 'max_power']].apply(remove, axis=1, result_type= 'expand')
```

```python
cars.head()
```

```python
cars.drop(['mileage','engine','max_power', 'torque'], axis= 1, inplace = True)	
```

```python
cars.head()
```

```python
cat_cols = ['fuel', 'seller_type','transmission', 'owner']
```

```python
dummies = pd.get_dummies(cars[cat_cols], drop_first = True)
```

```python
cars.drop(cat_cols, axis = 1,inplace = True)

cars.head()
```

```python
cars = pd.concat([cars,dummies],axis = 1)
```

```python
cars.head()
```

```python
x = cars.drop('selling_price', axis = 1)

y = cars['selling_price']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
num_cols = ['km_driven', 'seats', 'number of years', 'mileage(kmpl)', 'engine(cc)', 'max_power(bhp)']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.fit_transform(X_test[num_cols])
```

```python
X_train[num_cols]
```

```python
# let's convert X_train and y_train into numpy arrays

X_train = X_train.values
X_test = X_test.values
y_train = y_train.values
y_test = y_test.values
```

Let's create our models
Using Linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn import metrics
```

```python
linear = LinearRegression()

linear.fit(X_train, y_train)
```

```python
#prediction 

y_pred = linear.predict(X_test)
```

```python
mse = metrics.mean_squared_error(y_test, y_pred)
```

```python
print(mse)
```

```python
rmse = np.sqrt(mse)

print(rmse)

print('\n')

r2_linear =metrics.r2_score(y_test,y_pred)

print(r2_linear)
```
